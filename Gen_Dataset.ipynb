```python
import yfinance as yf
import pandas as pd
```

```python
def gen_dataset(ticker_pair, start_date, end_date):
    dataset = []
    for pair in ticker_pair:
        ticker_data = {}
        data = yf.download(tickers=pair, start=start_date, end=end_date)
        for col in data.columns:
            ticker_data['currency'] = [pair]*len(data)
            ticker_data['date'] = data[col].index
            ticker_data[col[0]] = data[col].values
        ticker_df = pd.DataFrame(ticker_data)
        
        dataset.append(ticker_df)

    dataset_df = pd.concat(dataset, ignore_index=True)

    return dataset_df
    
```

```python
start_date = '2015-01-05'
end_date = '2024-12-28'
pairs = ['EURUSD=X', 'VND=X', 'BTC-USD']
dataset_df = gen_dataset(ticker_pair=pairs, start_date=start_date, end_date=end_date)
dataset_df.loc[dataset_df['currency']=='EURUSD=X', 'currency'] = 'EUR-USD'
dataset_df.loc[dataset_df['currency']=='VND=X', 'currency'] = 'USD-VND'
dataset_df
```

```python
dataset_df.to_csv('yfinance_data.csv', index=False)
```
